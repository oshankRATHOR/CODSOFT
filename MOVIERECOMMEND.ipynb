```python
import ast

# Step 1: If needed, convert from string to list of dicts
def convert(obj):
    if isinstance(obj, str):
        try:
            parsed = ast.literal_eval(obj)
            return [i['name'] for i in parsed if isinstance(i, dict) and 'name' in i]
        except:
            return []
    elif isinstance(obj, list):
        return [i['name'] for i in obj if isinstance(i, dict) and 'name' in i]
    return []

# Apply conversion only if data is in JSON-like string format
for col in ['Genre', 'Director', 'Actor 1', 'Actor 2', 'Actor 3']:
    movies[col] = movies[col].apply(convert)

# Step 2: Remove spaces inside genre/actor/director names
for col in ['Genre', 'Director', 'Actor 1', 'Actor 2', 'Actor 3']:
    movies[col] = movies[col].apply(lambda x: [i.replace(" ", "") for i in x])

# Step 3: Create tags column
movies['tags'] = (
    movies['Genre'].apply(lambda x: ' '.join(x)) + ' ' +
    movies['Year'].astype(str) + ' ' +
    movies['Director'].apply(lambda x: ' '.join(x)) + ' ' +
    movies['Actor 1'].apply(lambda x: ' '.join(x)) + ' ' +
    movies['Actor 2'].apply(lambda x: ' '.join(x)) + ' ' +
    movies['Actor 3'].apply(lambda x: ' '.join(x))
)
```
